## Imports

```python
import os
import sys
sys.path.insert(1, os.path.realpath(os.path.pardir))
from bbtoolkit.preprocessing.environment import GeometryFactory, Geometry
from utils.environment import StandardBuildingGeometryProcessor, get_geometry_by_name
import numpy as np
```

## Create two-room geometry

```python
geometry_name = 'two_room'
cfg_path = f'../cfg/envs/{geometry_name}.ini'
geometry = GeometryFactory(
    cfg_path,
    get_geometry_by_name,
    StandardBuildingGeometryProcessor
)(
    getter_kwargs=dict(geometry_name=geometry_name),
    building_processor_kwargs=dict(geometry_name=geometry_name)
)
```

### Plot training space

```python
fig = geometry.training_space.plot()
```

### Save geometry

```python
geometry.save(f'../data/envs/{geometry_name}.pkl')
```

### Read geometry

```python
geometry = Geometry.load(f'../data/envs/{geometry_name}.pkl')
```
