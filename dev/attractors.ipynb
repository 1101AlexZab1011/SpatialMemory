```python
import numpy as np
import matplotlib.pyplot as plt

def create_circle_matrix(m, n, radius = None):
    # Create an m x n matrix filled with zeros
    matrix = np.zeros((m, n))
    if radius is None:
        radius = min(m, n) // 2

    # Calculate the center of the matrix
    # center_x, center_y = m // 2, n // 2
    center_x, center_y = m // 2, n // 2

    # Iterate over the matrix and fill in the circle
    for i in range(m):
        for j in range(n):
            # Calculate the distance from the center
            distance = np.sqrt((i - center_x) ** 2 + (j - center_y) ** 2)
            # If the distance is less than or equal to the radius, mark it as part of the circle
            if distance <= radius:
                matrix[i, j] = 1

    return matrix

# Example usage
m, n = 10, 10  # Dimensions of the matrix
radius = 3  # Radius of the circle
circle_matrix = create_circle_matrix(m, n, radius)
print(circle_matrix)
```

```python
import numpy as np

def create_gaussian_matrix(m, n, sigma=None):
    # Create an m x n matrix
    matrix = np.zeros((m, n))

    if sigma is None:
        sigma = min(m, n) // 10

    # Calculate the center of the matrix
    center_x, center_y = m // 2, n // 2

    # Iterate over the matrix to apply the Gaussian function
    for i in range(m):
        for j in range(n):
            # Calculate the Gaussian value
            matrix[i, j] = np.exp(-((i - center_x) ** 2 + (j - center_y) ** 2) / (2 * sigma ** 2))

    return matrix

# Example usage
m, n = 10, 10  # Dimensions of the matrix
sigma = 1.5  # Standard deviation for the Gaussian distribution
# gaussian_matrix = create_gaussian_matrix(m, n, sigma)
# print(gaussian_matrix)
```

```python
import numpy as np

def create_ricker_matrix(m, n, a=None):
    # Create an m x n matrix
    matrix = np.zeros((m, n))

    if a is None:
        a = min(m, n) // 10

    # Calculate the center of the matrix
    center_x, center_y = m // 2, n // 2

    # Iterate over the matrix to apply the Ricker wavelet function
    for i in range(m):
        for j in range(n):
            # Calculate distances from the center
            x = i - center_x
            y = j - center_y
            # Calculate the Ricker wavelet value
            factor = (x**2 + y**2) / a**2
            matrix[i, j] = (1 - factor) * np.exp(-factor / 2)

    return matrix

# Example usage
m, n = 10, 10  # Dimensions of the matrix
a = 3  # Scale parameter for the Ricker wavelet
ricker_matrix = create_ricker_matrix(m, n, a)
print(ricker_matrix)
```

```python
import numpy as np

def create_index_matrix(m, n):
    # Calculate the center indices
    center_x, center_y = m // 2, n // 2

    # Generate a matrix for x-axis indices
    x_indices = np.arange(m).reshape(m, 1) - center_x
    # Generate a matrix for y-axis indices
    y_indices = np.arange(n).reshape(1, n) - center_y

    # Tile the indices to form matrices of shape (m, n)
    x_matrix = np.tile(x_indices, (1, n))
    y_matrix = np.tile(y_indices, (m, 1))

    return x_matrix, y_matrix

# Example usage
m, n = 5, 5  # Dimensions of the matrix
x_matrix, y_matrix = create_index_matrix(m, n)
print("X Indices Matrix:")
print(x_matrix)
print("\nY Indices Matrix:")
print(y_matrix)
```

```python
def wrap_indices(i, j, rows, cols):
    """
    Wraps the indices i (row) and j (column) for a matrix with dimensions (rows, cols).

    Args:
        i (int): The row index.
        j (int): The column index.
        rows (int): The number of rows in the matrix.
        cols (int): The number of columns in the matrix.

    Returns:
        tuple: The wrapped row and column indices.
    """
    # Wrap the row index
    wrapped_i = ((i % rows) + rows) % rows
    # Wrap the column index
    wrapped_j = ((j % cols) + cols) % cols
    return (wrapped_i, wrapped_j)

# Example usage
rows, cols = 10, 10  # Example matrix dimensions
i, j = -2, 12  # Example indices
wrapped_indices = wrap_indices(i, j, rows, cols)
print(f"Wrapped indices for i={i}, j={j}: {wrapped_indices}")
```

```python
import numpy as np

class AttractorIndexer:
    def __init__(self, shape):
        self.shape = shape

    def __getitem__(self, shift):
        vertical_shift, horizontal_shift = shift
        vertical_shift *= -1
        rows, cols = self.shape

        # Calculate new row indices
        row_indices = (np.arange(rows) - vertical_shift) % rows
        # Calculate new column indices
        col_indices = (np.arange(cols) - horizontal_shift) % cols

        return np.ix_(row_indices, col_indices)
```

```python

class AttractorIndexer:
    def __init__(self, shape):
        self.shape = shape

    def __getitem__(self, shift):
        vertical_shift, horizontal_shift = shift
        vertical_shift *= -1
        rows, cols = self.shape

        # Calculate new row indices
        row_indices = (np.arange(rows) - vertical_shift) % rows
        # Calculate new column indices
        col_indices = (np.arange(cols) - horizontal_shift) % cols

        return np.ix_(row_indices, col_indices)


class AttractorState:
    def __init__(self, kernel: np.ndarray, weights: np.ndarray, indexer: AttractorIndexer):
        self.kernel = kernel
        self.weights = weights
        self.indexer = indexer

    @property
    def shape(self):
        return self.kernel.shape

    def __getitem__(self, indices):
        i, j = wrap_indices(*indices, *self.kernel.shape)
        ratio = self.kernel[i, j]
        return self.weights[self.indexer[indices]]*ratio

    def __repr__(self):
        return f"AttractorState(attractor={self.kernel}, weights={self.weights})"


class Attractor:
    def __init__(
        self,
        attractor_kernel,
    ):
        self.attractor_kernel = attractor_kernel
        self.attractor_weights = attractor_kernel[create_index_matrix(*attractor_kernel.shape)]
        self.indexer = AttractorIndexer(attractor_kernel.shape)

    @property
    def shape(self):
        return self.attractor_kernel.shape

    def __call__(self, weights: np.ndarray) -> AttractorState:
        return AttractorState(self.attractor_weights, weights, self.indexer)

attr = Attractor(create_circle_matrix(6, 6, 1))
state = attr(np.arange(36).reshape(6, 6))

state[0, 1]


```
